```python
import pandas as pd
from scipy.stats import ttest_ind

# 1. Load Data
path = r'D:\@02 Personal\@03 Mini Data Project\@07 DEC Portfolio\data\test1_menu.xlsx'
df = pd.read_excel(path)

# 2. Pisahkan data berdasarkan varian
group_a = df[df['variant'] == 'A_horizontal_menu']
group_b = df[df['variant'] == 'B_dropdown_menu']

# 3. Tentukan metrik yang ingin diuji
metrics = ['revenue', 'pages_viewed', 'added_to_cart', 'bounced']
alpha_original = 0.05

# 4. Hitung Bonferroni Correction (Alpha Baru)
# Rumus: Alpha / Jumlah Metrik
alpha_adjusted = alpha_original / len(metrics)

results = []

for m in metrics:
    # Lakukan T-Test untuk setiap metrik
    t_stat, p_val = ttest_ind(group_a[m].dropna(), group_b[m].dropna())
    
    # Tentukan apakah lolos setelah koreksi
    status = "✅ SIGNIFIKAN" if p_val < alpha_adjusted else "❌ TIDAK SIGNIFIKAN"
    
    results.append({
        'Metric': m,
        'Mean A': group_a[m].mean(),
        'Mean B': group_b[m].mean(),
        'P-Value Raw': p_val,
        'Verdict': status
    })

# Tampilkan Hasil
df_results = pd.DataFrame(results)
print(f"=== HASIL UJI SIGNIFIKANSI (Bonferroni Adjusted Alpha: {alpha_adjusted:.4f}) ===")
print(df_results[['Metric', 'Mean A', 'Mean B', 'P-Value Raw', 'Verdict']])
```
