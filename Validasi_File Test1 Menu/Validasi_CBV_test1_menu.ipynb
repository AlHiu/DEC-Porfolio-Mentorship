```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. BACA FILE (Wajib diulangi jika Anda memulai sesi baru)
path = r'D:\@02 Personal\@03 Mini Data Project\@07 DEC Portfolio\data\test1_menu.xlsx'
df = pd.read_excel(path)

# 2. HITUNG DISTRIBUSI (COVARIATE BALANCE)
# Kita membuat tabel silang antara Varian dan Device Type
# normalize='index' mengubah angka absolut menjadi persentase 0-1
balance_table = pd.crosstab(df['variant'], df['device_type'], normalize='index') * 100

print("=== PERSENTASE DISTRIBUSI PERANGKAT ===")
print(balance_table.round(2))

# 3. VISUALISASI
# Menggunakan bar chart untuk melihat apakah tinggi kotaknya setara
balance_table.plot(kind='bar', figsize=(8, 5))

plt.title('Pemeriksaan Keseimbangan: Varian vs Perangkat', fontsize=12)
plt.ylabel('Persentase (%)')
plt.xlabel('Varian Menu')
plt.xticks(rotation=0)
plt.legend(title='Device Type')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
