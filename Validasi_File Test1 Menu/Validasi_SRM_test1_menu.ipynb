```python
import pandas as pd
from scipy.stats import chisquare

# 1. Load Data
path = r'D:\@02 Personal\@03 Mini Data Project\@07 DEC Portfolio\data\test1_menu.xlsx'
df = pd.read_excel(path)

# 2. Menghitung distribusi berdasarkan kemunculan baris (Raw Counts)
# Kita gunakan ini karena filter sebelumnya menghapus seluruh data Anda
counts = df['variant'].value_counts()
observed = counts.values
variant_names = counts.index.tolist()

# 3. Hitung Expected (asumsi 50:50)
total_data = sum(observed)
expected = [total_data / 2, total_data / 2]

# 4. Uji Chi-Square
chi_stat, p_value = chisquare(observed, f_exp=expected)

print("=== HASIL ANALISIS SRM (RAW DATA) ===")
print(f"Total Baris: {total_data}")
for name, val in zip(variant_names, observed):
    print(f"- {name}: {val} baris")

print("-" * 30)
print(f"P-Value: {p_value:.10f}")

if p_value < 0.001:
    print("ðŸš¨ STATUS: SRM TERDETEKSI!")
    print("Jumlah data antar varian sangat jomplang secara statistik.")
else:
    print("âœ… STATUS: TIDAK TERDETEKSI SRM.")
    print("Distribusi data antar varian seimbang.")
```
